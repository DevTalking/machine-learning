```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)
x
```

```python
siny = np.sin(x)
siny
```

```python
plt.plot(x, siny)
plt.show()
```

```python
cosy = np.cos(x)
plt.plot(x, cosy)
plt.show()
```

```python
plt.plot(x, cosy)
plt.plot(x, siny)
plt.show()
```

```python
plt.plot(x, cosy, color="red")
plt.plot(x, siny)
plt.show()
```

```python
plt.plot(x, cosy, linestyle='--')
plt.plot(x, siny)
plt.show()
```
